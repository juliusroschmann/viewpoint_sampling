```python
import xml.etree.ElementTree as ET
import pandas as pd

# Read the xml file
tree = ET.parse('Marker_6x6.xml')
root = tree.getroot()

# Create a list of data points
data_points = []
size = 2.5
# Iterate through each marker
for marker_idx, marker in enumerate(root.findall('marker')):

    # Get the marker index
    index = int(marker.attrib.get('index'))
    # print("Marker_idx: ", marker_idx)
    # Iterate through each corner in the marker
    for corner_idx, corner in enumerate(marker.findall('corner')):

        # Get the x, y and z coordinates for the corner
        xo = float(corner.attrib.get('x'))
        yo = float(corner.attrib.get('y'))
        zo = float(corner.attrib.get('z'))

        z = 0.0

        if corner_idx == 0:
        # centerpoint
            #print("center_points: ", center_points[marker_idx])
            fx = xo/9
            fy = yo/9
            #print("fx: {}, fy: {}".format(fx, fy))
            # x = center_points[marker_idx][0]
            # y = center_points[marker_idx][1]
            x = fx*5
            y = fy*5
            data_points.append([index, x, y, z])

        elif corner_idx == 1:
            xx = x + size
            yy = y - size
            data_points.append([index, xx, yy, z])
        
        elif corner_idx == 2:
            xx = x + size
            yy = y + size
            data_points.append([index, xx, yy, z])    

        elif corner_idx == 3:
            xx = x - size
            yy = y + size
            data_points.append([index, xx, yy, z])
        # Append the data points to the list
        

# Create a pandas dataframe
df = pd.DataFrame(data_points, columns=['index','x','y','z'])

# Write the dataframe to a new xml file
new_xml_root = ET.Element('multimarker', markers="36")

# Iterate through each marker in the dataframe
for index, marker in df.groupby('index'):
    new_marker = ET.SubElement(new_xml_root, 'marker', index=str(index), status="1")
    for corner in marker.values:
        ET.SubElement(new_marker, 'corner', x=str(corner[1]), y=str(corner[2]), z=str(corner[3]))

tree = ET.ElementTree(new_xml_root)
tree.write('new_6x6.xml')
df.head(1000)

# print(df["index"]==0)
```
